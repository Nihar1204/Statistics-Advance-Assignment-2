```python
# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

# Probability Mass Function (PMF):
# The PMF is used for discrete random variables, which are variables that can only take on specific, isolated values.
# It gives the probability that a discrete random variable is exactly equal to some value.
# The PMF is represented as a function that maps each possible value of the random variable to its probability.
# Mathematically, if X is a discrete random variable, the PMF of X, denoted as P(X), is defined as P(X = x), where x 
# is a specific value of X.
# Example:
# Consider the roll of a fair six-sided die. The PMF for this scenario would assign probabilities to each possible outcome 
# (1, 2, 3, 4, 5, or 6). Since each outcome has an equal chance of occurring (assuming the die is fair), the PMF for this 
# situation would be:P(X = 1) = 1/6,... , P(X = 6) = 1/6.

# Probability Density Function (PDF):
# The PDF is used for continuous random variables, which are variables that can take on any value within a certain range.
# Unlike the PMF, which assigns probabilities to specific values, the PDF gives the relative likelihood that a continuous
# random variable falls within a particular interval.
# The PDF is represented as a function that describes the distribution of probabilities over the entire range of possible
# values.
# The probability of a continuous random variable falling within a specific interval is given by the area under the PDF 
# curve over that interval.
# Example:
# Consider the height of adult males in a certain population. The PDF for this scenario might follow a normal (Gaussian)
# distribution, with most heights clustering around the mean. The PDF curve would describe the likelihood of observing 
# various heights within the population.
```

```python
# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

# The Cumulative Distribution Function (CDF) is a concept in probability theory that provides the cumulative probability
# distribution of a random variable. It gives the probability that a random variable X is less than or equal to a certain 
# value x. The CDF is a useful tool for understanding the distribution of a random variable and for calculating 
# probabilities associated with it.
# Example:
# Consider the roll of a fair six-sided die. Let X be the random variable representing the outcome of the roll. The CDF 
# F(x) for this scenario would be:
# F(x) = Probability that the outcome of the roll is less than or equal to ùë•
# Since each outcome of the die roll is equally likely, the CDF can be calculated as follows:
# For x=1: F(1) = F(X<=1) = 1/6, For x=2: F(2) = F(X<=2) = 2/6 = 1/3 ... For x=6: F(6) = F(X<=6) = 6/6 = 1
# So, the CDF F(x) for the roll of a fair six-sided die would be a step function that increases by 1/6 at each integer
# value of x from 1 to 6.

# Uses:
# Probability Calculation: The CDF allows for easy calculation of probabilities associated with a random variable. By using 
# the CDF, we can quickly find the probability that a random variable falls within a certain range.
# Statistical Analysis: CDFs are often used in statistical analysis to assess the characteristics of a dataset, such
# as its central tendency and variability.
```

```python
# Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
# is normally distributed, what is the probability that a randomly selected observation will be greater
# than 60? Use the appropriate formula and show your calculations.

# We can use the standard normal distribution (Z-distribution) and then we can refer the Z-table.
# Here we have mean = 50, std = 10, x = 10.
# z = (x - mean)/std = (60 - 50)/10 = 1
# Now, we need to find the probability that a randomly selected observation will be greater than 60. Since we're
# interested in the area to the right of 60 on the standard normal distribution curve, we look for the probability 
# corresponding to Z=1.
# Using a Z-table we find that the probability corresponding to Z=1 is approximately 0.8413.
# So, the probability that a randomly selected observation will be greater than 60 is approximately 0.8413 or 84.13%.
```

```python
# Q8: What is the z score? State the importance of the z score.

# The Z-score, also known as the standard score, is a statistical measure that quantifies the number of standard 
# deviations a data point is from the mean of a dataset. It indicates how many standard deviations an observation 
# is above or below the mean of the dataset.
# Formula:
# z = (x - mean)/std

# Importance of Z-score:
# Standardization: The Z-score standardizes data, allowing for comparison of data points from different distributions.
# By converting raw data into Z-scores, we can compare data points that are measured using different units or scales.
# Identifying Outliers: Z-scores help identify outliers in a dataset. Data points with Z-scores that are significantly 
# higher or lower than the mean may indicate unusual or extreme values.
# Probability Calculation: Z-scores are used in probability calculations and statistical hypothesis testing. In many 
# statistical tests, Z-scores are compared to critical values to determine statistical significance.
# Data Transformation: Z-scores are often used in data transformation techniques, such as normalization and 
# standardization, which are commonly used in machine learning algorithms and data preprocessing.
# Visual Representation: Z-scores provide a standardized way to visualize and interpret data distributions. Z-score
# plots and histograms allow for easy comparison and interpretation of data distributions.
```

```python
# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

# Central Limit Theorem states the sampling distribution of the sample mean approaches a normal distribution as the 
# sample size increases, regardless of the shape of the population distribution.

# Significance of the Central Limit Theorem:
# Statistical Inference: The Central Limit Theorem is the foundation of many statistical methods and techniques,
# including hypothesis testing, confidence intervals, and parameter estimation. It allows statisticians to make 
# probabilistic statements about sample means and population parameters based on sample data.
# Sampling Design: The Central Limit Theorem guides sampling design by providing insights into the behavior of sample
# means. It helps researchers determine appropriate sample sizes to achieve desired levels of precision and confidence
# in their estimates.
# Normal Approximation: The Central Limit Theorem allows for the approximation of various discrete distributions by
# the normal distribution. This simplifies calculations and facilitates statistical analysis in many situations.
```

```python
# Q10: State the assumptions of the Central Limit Theorem.

# Independent and Identically Distributed (i.i.d.) Samples: The samples drawn from the population must be independent
# of each other and have the same probability distribution. This means that each observation in the sample is not 
# influenced by any other observation, and all observations are drawn from the same underlying population distribution.
# Finite Variance: The population from which the samples are drawn must have a finite variance. This ensures that the 
# sample means have finite variance as well. If the population has an infinite variance, the Central Limit Theorem may 
# not hold.
# Large Sample Size: The Central Limit Theorem works best when the sample size (n) is sufficiently large. While there 
# is no strict cutoff for what constitutes a "large" sample size, a common rule of thumb is that the sample size should
# be at least 30. As the sample size increases, the distribution of sample means approaches a normal distribution more
# closely.
# Random Sampling: The samples must be drawn randomly from the population, without any bias. Random sampling ensures 
# that the sample is representative of the population and that the results can be generalized to the population as a
# whole.
# Population Distribution: While the Central Limit Theorem is often stated without specifying the shape of the
# population distribution, it tends to hold true for a wide range of population distributions, including non-normal
# distributions. However, the theorem may not work well for extremely skewed or heavy-tailed distributions.
```
